### Exercício 2 

Função 1

$$(a) ~~ f(x_{1}, x_{2})= x_{1}^{2}-3x_{1}x_{2}+4x_{2}^{2}+x_{1}-x_{2}$$

- Ponto inicial: $x^{0}=\{1,2\}^{t}$
- Direção: $d = \{-1, -2\}^{t}$ 

```python
def func(x):
    return x[0]**2 - 3*x[0]*x[1] + 4*x[1]**2 + x[0] - x[1]  

initial_point = np.array([1, 2])
direction = np.array([-1, -2])

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = func([X, Y])
```

```python
constant_st = constant_step(func, initial_point, direction)
plot_optimization(X, Y, Z, constant_st, path = './figures/constant_step_2a.pdf')
```

```python
bisec = bisection(func, initial_point, direction)
plot_optimization(X, Y, Z, bisec, path='./figures/bisection_2a.pdf')
```

```python
golden_sec = golden_section(func, initial_point, direction)
plot_optimization(X, Y, Z, golden_sec, path='./figures/golden_section_2a.pdf')
```

Função de MacCormick

$$(b) ~~ f(x_{1}, x_{2}) = sin(x_{1}+x_{2}) + (x_{1}-x_{2})^{2}-1.5x_{1}+2.5x_{2}$$

- Ponto inicial: $x^{0} = \{-2,3\}^{t}$
- Direção: $d = \{1.453, -4.547\}^{t}$

```python
def func(x):
    return np.sin(x[0]+x[1]) + (x[0]-x[1])**2-1.5*x[0]+2.5*x[1] 

initial_point = np.array([-2, 3])
direction = np.array([1.453, -4.547])

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = func([X, Y])
```

```python
constant_st = constant_step(func, initial_point, direction)
plot_optimization(X, Y, Z, constant_st, path='./figures/constant_step_2b.pdf')
```

```python
bisec = bisection(func, initial_point, direction)
plot_optimization(X, Y, Z, bisec, path='./figures/bisection_2b.pdf')
```

```python
golden_sec = golden_section(func, initial_point, direction)
plot_optimization(X, Y, Z, golden_sec, path='./figures/golden_section_2b.pdf')
```

Função de Himmelblau

$$(c) ~~ f(x_{1}, x_{2}) = (x_{1}^{2}+x_{2}-11)^{2} + (x_{1}+x_{2}^{2}-7)^{2}$$
