### Exercício 2 
$$(a) ~~ f(x_{1}, x_{2})= x_{1}^{2}-3x_{1}x_{2}+4x_{2}^{2}+x_{1}-x_{2}$$

- Ponto inicial: $x^{0}=\{1,2\}^{t}$
- Direção: $d = \{-1, -2\}^{t}$ 

```python
def func(x):
    return x[0]**2 - 3*x[0]*x[1] + 4*x[1]**2 + x[0] - x[1]  
```

```python
l_bound = np.array([-10, -10])
u_bound = np.array([10, 10])

golden_sec = golden_section(function=func, l_bound=l_bound, u_bound=u_bound)
golden_sec
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = func([X, Y])

plot_golden_section(X, Y, Z, golden_sec)
```

Função de MacCormick

$$(b) ~~ f(x_{1}, x_{2}) = sin(x_{1}+x_{2}) + (x_{1}-x_{2})^{2}-1.5x_{1}+2.5x_{2}$$

- Ponto inicial: $x^{0} = \{-2,3\}^{t}$
- Direção: $d = \{1.453, -4.547\}^{t}$

```python
def func(x):
    return np.sin(x[0]+x[1]) + (x[0]-x[1])**2-1.5*x[0]+2.5*x[1] 
```

```python
l_bound = np.array([-10, -10])
u_bound = np.array([10, 10])

golden_sec = golden_section(function=func, l_bound=l_bound, u_bound=u_bound)
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = func([X, Y])

plot_golden_section(X, Y, Z, golden_sec)
```

Função de Himmelblau

$$(c) ~~ f(x_{1}, x_{2}) = (x_{1}^{2}+x_{2}-11)^{2} + (x_{1}+x_{2}^{2}-7)^{2}$$

- Ponto inicial: $x^{0}=\{0,5\}^{t}$
- Direção: $d=\{3,1.5\}^{t}$

```python
def func(x):
    return (x[0]**2+x[1]-11)**2 + (x[0]+x[1]**2-7)**2
```

```python
l_bound = np.array([-10, -10])
u_bound = np.array([10, 10])

golden_sec = golden_section(function=func, l_bound=l_bound, u_bound=u_bound)
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = func([X, Y])

plot_golden_section(X, Y, Z, golden_sec)
```
