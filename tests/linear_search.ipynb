```python
import sys
sys.path.append('../')

import time
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from qsopt.search import line_search
```

```python
def f(x):
    return x[0]**2 - 3*x[0]*x[1] + 4*x[1]**2 + x[0] - x[1]  

x0 = np.array([1, 2])
d = np.array([-1, -2])
lb = np.array([-10, -10])
ub = np.array([10, 10])
method = line_search(f, x0, d)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = f([X, Y])

fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(12, 5))

fig.delaxes(ax[0])
ax[0] = fig.add_subplot(121, projection='3d')
ax[0].view_init(40, 20)
ax[0].plot_surface(X, Y, Z, cmap='viridis', linewidth =0)
ax[0].set_xlabel('x1')
ax[0].set_ylabel('x2')
ax[0].set_zlabel(r"$f(x_{1}, x_{2})$")
ax[0].set_title(r"Superfície da função $f(x_{1}, x_{2})$")

contour = ax[1].contour(X, Y, Z, 200, cmap='viridis')
ax[1].scatter(method['xopt'][0], method['xopt'][1], color='red', s=50, marker='o', label=f'Ponto Mínimo - num_iter: {method["num_iter"]}')
ax[1].text(method['xopt'][0] + 1, method['xopt'][1] - 2, f'{method["xopt"][0]}, \n{method["xopt"][1]}', size=10, zorder=2, color='k', bbox=dict(boxstyle="round", alpha=1, facecolor='white'))
for i in range(len(method['path_a'])):
    ax[1].plot(method['path_a'][i], method['path_a'][i], 'red', marker='x')
    ax[1].plot(method['path_b'][i], method['path_b'][i], 'blue', marker='x')
ax[1].set_xlabel('x1')
ax[1].set_ylabel('x2')
ax[1].legend(loc='upper right')

plt.show()
```
