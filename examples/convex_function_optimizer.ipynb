```python
import sys
sys.path.append('../')

import numpy as np
from optymus.opt import Optimizer
from optymus.utils import mccormick_function
```

```python
f = mccormick_function()
initial_point = np.array([2,2])
```

```python
optimizer = Optimizer(f_obj=f,
                      x0=initial_point,
                      method='bfgs')
```

```python
optimizer.report()
```

```python
optimizer.plot(print_opt=False)
```
