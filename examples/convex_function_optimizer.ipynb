```python
import sys
sys.path.append('../')

import jax.numpy as jnp
from optymus.optim import Optimizer
from optymus.utils import sphere_function
```

```python
f = sphere_function()
initial_point = jnp.array([2.0, 2.0])
```

```python
optimizer = Optimizer(f_obj=f,
                      x0=initial_point,
                      method='adagrad',
                      )
```

```python
optimizer.report()
```

```python
optimizer.plot(print_opt=False)
```

```python
# Run Adam optimization
from optymus.optim._adaptative import adam
result = adam(f, initial_point)

print("Optimal point:", result['xopt'])
print("Minimum value:", result['fmin'])
print("Number of iterations:", result['num_iter'])
```
